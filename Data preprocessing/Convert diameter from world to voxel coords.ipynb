```python
import SimpleITK as sitk
import numpy as np
import pandas as pd
import csv
import os
```

```python
# Set data split to 'train', 'test', or 'validation'

data_split = 'validation'
csv_path = f'candidates_{data_split}.csv'

df = pd.read_csv(csv_path)
df.head()
```

```python
# Note- the function requires that you have ran the cell above!

def convert_diameter(index):
    
    seriesuid = df.loc[index, 'seriesuid']
    filename = f'{data_split}_data_raw/{seriesuid}.mhd'
    itkimage = sitk.ReadImage(filename)
    spacing = np.array(list(reversed(itkimage.GetSpacing())))[2]
    
    d_mul = df.loc[index, 'diameter_mm']*spacing
    d_div = df.loc[index, 'diameter_mm']/spacing
    
    return d_mul, d_div
```

```python
for i in range(df.shape[0]):
    df.loc[i, 'voxel_d'] = convert_diameter(i)[0]
    df.loc[i, 'voxel_d2'] = convert_diameter(i)[1]
```

```python
df.to_csv(csv_path)
```
